## Functions

```python
def shaldon_knocks(name,number_of_times=4):
    for i in range(number_of_times):
        print("Knock Knock Knock {}".format(name))
```

```python
shaldon_knocks("Penny",10)
```

# Try Catch

```python
def div(a,b):
    try:
        return a/b
    except:
        print("Error Comes")
    finally:
        print("Wrapping Up")
        # return 10         # If finally has return value it will not see try return value
```

```python
ans = div(10,15)
print(ans)
```

```python
ans = div(10,15)
print(ans)
```

# Local and Global Variables

```python
x = 10
def main():
    x = 5
    print(x) # accessing global variable
```

```python
main()
print(x)
```

```python
x = 10
def main():
    global x
    x += 5  # this is only applicable when we apply global to local
    print(x) # accessing global variable
```

```python
main()
```

```python
def outer():
    x = "local"

    def inner():
        print(x)

    inner()
    print(x)
```

```python
# del x
outer()
```

```python
def outer():
    x = 10

    def inner():
        nonlocal x # if x is present in enclosure
        x+=5
        print(x)

    inner()
    print(x)
```

```python
outer()
```
