# Try Catch

# Local and Global Variables

```python
x = 10
def main():
    x = 5
    print(x)
```

```python
main()
print(x)
```

```python
x = 10
def main():
    global x
    x += 5  # this is only applicable when we apply global to local
    print(x) # accessing global variable
```

```python
main()
```

```python
def outer():
    x = "local"

    def inner():
        print(x)

    inner()
    print(x)
```

```python
# del x
outer()
```

```python
def outer():
    x = 10

    def inner():
        nonlocal x # if x is present in enclosure
        x+=5
        print(x)

    inner()
    print(x)
```

```python
outer()
```

```python
# Create an Object in Py
users = {
    "anurag":"password",
    "someone":"onlymine"
}
```

```python
def show(username,password):
    if username in users and users[username] == password:
        print("Hello World")
    else:
        print("Not Authenticated")
```

```python
show("anurag","password")
```

```python
def sum(username,password,a,b):
    if username in users and users[username] == password:
        print("The Sum is {}".format(a+b))
    else:
        print("Not Authanticated")
```

```python
sum("anurag","password",56,12)
```

```python
users = {
    "anurag": "password",
    "raj": "12345"
}

def logger_function(func):
    def wrapper(username, password, *args, **kwargs):
        if username in users and users[username] == password:
            return func(*args, **kwargs)
        else:
            print("Not Authenticated")
    return wrapper

@logger_function
def add(a, b):
    print(a + b)

add("anurag", "password", 3, 5)
```
